```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.xlabel('Значения X') 
plt.ylabel('Занчения Y')

# Установка границ
ax = plt.gca()
ax.spines['top'].set_visible(False)  #скрыли
ax.spines['right'].set_visible(False) #скрыли
ax.spines['bottom'].set_linewidth(2) #утощение
ax.spines['left'].set_linewidth(2) #утолщение
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 8, 27, 64, 125]

fig, ax = plt.subplots()

# График 1
plt.plot(x, y1, label='График 1')

# График 2
plt.plot(x, y2, label='График 2')

# Настройка осей и легенды
ax.set_xlabel('Значения X') 
ax.set_ylabel('Занчения Y')
ax.set_title('График заивисимости занчений X от занчений Y')
ax.legend()

# Включение сетки
ax.grid(True)

# Добавление текстовой аннотации
ax.annotate('Точка 1', (1, 1), (1.5, 20), arrowprops={'arrowstyle' : '->'})

# Добавление минорных делений на осях
ax.minorticks_on()
ax.tick_params(which = 'minor', axis = 'both', width = 2)

# Установка границ
ax = plt.gca()
ax.spines['top'].set_visible(False)  #скрыли
ax.spines['right'].set_visible(False) #скрыли
ax.spines['bottom'].set_linewidth(0.5) #утощение
ax.spines['left'].set_linewidth(0.5) #утолщение
plt.show()
```

Stackplot

```python
np.random.seed(234234) # Фиксируем seed для воспризводимости 

rng = np.arange(50) 
rnd = np.random.randint(0, 10, size = (3, rng.size))  # Матрица 3х50 из чисел от 0 до 10
yrs = 1950 + rng # Массив годов

fig, ax = plt.subplots(figsize = (5,3))

ax.stackplot(yrs, rng + rnd, labels = ['Eastasia', 'Eurasia', 'Oceania'])
ax.set_title('Combined debt growth over time')
ax.legend(loc = 'upper left')
ax.set_ylabel('Total debt')
ax.set_xlim(xmin = yrs[0], xmax = yrs[-1])

fig.tight_layout() # Делает отображение элементов контейнера более компактным
fig.show()
```

scatterplot и hist

```python
x = np.random.randint(low=1, high=11, size=50)
y = x + np.random.randint(1, 5, size=x.size)
data = np.column_stack((x, y))

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

ax1.scatter(x=x, y=y, marker='o', c='b', ) # scatter plot
ax1.set_title('Scatter: x versus y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.hist(data, bins=np.arange(data.min(), data.max()), label=('x', 'y'))
ax2.legend(loc=(0.65, 0.8))
ax2.set_title('Frequencies of x and y')
ax2.yaxis.tick_right()

plt.show()
```

```python
x1 = [1, 1, 1, 3, 4, 3]
y1 = [3, 3, 4, 5, 5, 6]
data1 = np.column_stack((x1, y1))

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

ax1.scatter(x=x1, y=y1, marker='x', c='r', ) # scatter plot
ax1.set_title('Scatter: x1 versus y1')
ax1.set_xlabel('x1')
ax1.set_ylabel('y1')

ax2.hist(data1, bins=np.arange(data1.min(), data1.max()+2), label=('x1', 'y1'))
ax2.legend(loc=(0.8, 0.8))
ax2.set_title('Frequencies of x1 and y1')
ax2.yaxis.tick_right()

plt.show()
```

```python
# auxiliary fuction
def add_titlebox(ax, text: str):
    ax.text(.55, .8, text,
        horizontalalignment='center',
        transform=ax.transAxes,
        bbox=dict(facecolor='white', alpha=0.6),
        fontsize=12.5)
    return ax

from io import BytesIO
import tarfile
from urllib.request import urlopen

url = 'http://www.dcc.fc.up.pt/~ltorgo/Regression/cal_housing.tgz'
b = BytesIO(urlopen(url).read())
fpath = 'CaliforniaHousing/cal_housing.data'

with tarfile.open(mode='r', fileobj=b) as archive:
    housing = np.loadtxt(archive.extractfile(fpath), delimiter=',')

y = housing[:, -1]
pop, age = housing[:, [4, 7]].T
```

Несколько подграфиков в одном графике

```python
gridsize = (3, 2)
fig = plt.figure(figsize=(12,8))
ax1 = plt.subplot2grid(gridsize, (0,0), colspan=2, rowspan=2)
ax2 = plt.subplot2grid(gridsize, (2, 0))
ax3 = plt.subplot2grid(gridsize, (2, 1))

ax1.set_title('Home value as a function of home age & area population', fontsize=14)
sctr = ax1.scatter(x=age, y=pop, c=y, cmap='RdYlGn')
plt.colorbar(sctr, ax=ax1, format='$%d')
ax1.set_yscale('log')   # Логарифмическая шкала - степени 10-ки

ax2.hist(age, bins='auto')

ax3.hist(pop, bins='auto', log=True)

add_titlebox(ax2, 'Histogram: home age')
add_titlebox( ax3, 'Histogram: area population (log scl.)')

plt.subplots_adjust(hspace=0.4, wspace=0.4) # Регулируем расстояния между подграфиками

plt.show()
```
